```python
from datetime import datetime
from docx import Document
from pathlib import Path
import pandas as pd
```

```python
def get_df(table):
    num_row = len(table.rows)
    num_col = len(table.rows[0].cells)
    
    header = table.rows[0].cells
    columns = [item.text for item in header]
    
    data = []
    for i in range(1, num_row):
        row = []
        for j in range(num_col):
            text = table.rows[i].cells[j].text
            row.append(text)
        data.append(row)
    
    df = pd.DataFrame(data, columns=columns)
    return df
```

```python
def get_date_string(row):
    year = 2023
    month = datetime.strptime(row['月份'], '%b').month
    day = int(row['号'])
    week = row['周']
    
    dt = datetime(year, month, day)
    date_string = dt.strftime('%d/%m/%Y') + f' ({week})'
    return date_string
```

```python
file_path = Path.cwd() / 'data.docx'
doc = Document(file_path)
table = doc.tables[0]
df = get_df(table)
```

```python
test_df = df[df['节目'].apply(lambda row: '启发课程' in row)].copy()

test_df['日期'] = test_df.apply(get_date_string, axis='columns')
test_df['建议歌曲'] = test_df['建议歌曲'].apply(lambda row: row.replace('\n', ' '))
test_df['伴奏'] = test_df.apply(lambda row: row['司琴'] if row['吉他'] == '-' else row['吉他'], axis='columns')

items = ['日期', '节目', '破冰', '领唱', '伴奏', '建议破冰', '建议歌曲']
test_df = test_df.filter(items)
```

```python
result = test_df.to_dict('records')
```

```python
for item in result:
    print(f"*{item['日期']}*")
    print(f"{item['节目']}")
    print(f"破冰: {item['破冰']}")
    print(f"领唱: {item['领唱']}")
    print(f"伴奏: {item['伴奏']}")
    print(f"建议破冰: {item['建议破冰']}")
    print(f"建议歌曲: {item['建议歌曲']}")
    print()
```

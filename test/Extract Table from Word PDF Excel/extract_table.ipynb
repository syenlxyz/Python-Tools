# Extract Table from Word

```python
from docx import Document
from pathlib import Path
import pandas as pd
```

```python
def get_df(table):
    num_row = len(table.rows)
    num_col = len(table.rows[0].cells)
    
    header = table.rows[0].cells
    columns = [item.text for item in header]
    
    data = []
    for i in range(1, num_row):
        row = []
        for j in range(num_col):
            text = table.rows[i].cells[j].text
            row.append(text)
        data.append(row)
    
    df = pd.DataFrame(data, columns=columns)
    return df
```

```python
file_path = Path.cwd() / 'data.docx'
doc = Document(file_path)
table = doc.tables[0]
df = get_df(table)
```

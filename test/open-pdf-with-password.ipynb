```python
from pypdf import PdfReader, PdfWriter
```

```python
from pathlib import Path
```

```python
file_path = Path.cwd() / '675W980305.pdf'
```

```python
reader = PdfReader(file_path, password='675W980305')
```

```python
writer = PdfWriter()
```

```python
pages = list(reader.pages)
```

```python
for page in pages:
    writer.add_page(page)
```

```python
output_path = Path.cwd() / 'output.pdf'
```

```python
with open(output_path, 'wb') as file:
    writer.write(file)
```
